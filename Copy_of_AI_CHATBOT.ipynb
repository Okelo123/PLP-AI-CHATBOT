```python
# BRIAN JUMA

# Step 1: Bot name and personality
bot_name = "CryptoBuddy"
bot_personality = "friendly"

# Step 2: Greeting function
def greet_user():
    print(f"üëã Hello! I'm {bot_name}, your friendly crypto companion!")
    print("üöÄ I can help you with crypto prices, explain blockchain terms, and more.")
    print("üí¨ What would you like to know today?")

# Step 3: Get input from the user
def get_user_input():
    return input("You: ")

# Step 4: Main function to run the chatbot
def main():
    greet_user()
    user_message = get_user_input()
    print(f"{bot_name}: I'm still learning, but I'll do my best to help with '{user_message}'!")

# Run the chatbot
if __name__ == "__main__":
    main()
```

```python
# üìà Rule-Based Logic - Profitability Focus
# üë§ Coded by: [Carolyne Katilo]

#  Crypto dataset
crypto_db = {
    "Bitcoin": {
        "price_trend": "rising",
        "market_cap": "high",
        "energy_use": "high",
        "sustainability_score": 3/10
    },
    "Ethereum": {
        "price_trend": "stable",
        "market_cap": "high",
        "energy_use": "medium",
        "sustainability_score": 6/10
    },
    "Cardano": {
        "price_trend": "rising",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 8/10
    }
}

# Function for profitable recommendation
def recommend_profitable_crypto(data):
    for coin, info in data.items():
        if info["price_trend"] == "rising" and info["market_cap"] == "high":
            return f"{coin} is trending up and has a strong market presence. It might be a solid choice for investment! üöÄ"
    return "No highly profitable cryptocurrencies found at the moment."

#  Chat interaction
user_input = input("Ask me about profitable cryptos: ").lower()

if "profit" in user_input or "growth" in user_input:
    print(recommend_profitable_crypto(crypto_db))
else:
    print("Try asking about profitability or growth opportunities!")
```

```python
# ‚ôªÔ∏è Rule-Based Logic - Sustainability Focus
# üë§ Coded by: [Raphael Kang'eri]

# Function for sustainable recommendation
def recommend_sustainable_crypto(data):
    sustainable_coins = []
    for coin, info in data.items():
        if info["energy_use"] == "low" and info["sustainability_score"] > 7/10:
            sustainable_coins.append((coin, info["sustainability_score"]))

    if sustainable_coins:
        # Sort by highest sustainability score
        best_coin = max(sustainable_coins, key=lambda x: x[1])
        return f"üå± {best_coin[0]} is the most sustainable choice with a score of {int(best_coin[1]*10)}/10! It's a great pick for eco-conscious investing."
    else:
        return "‚ö†Ô∏è No crypto meets the top sustainability criteria at the moment."

# Chat interaction for sustainability
user_input_sustainability = input("Ask me about green or eco-friendly cryptos: ").lower()

if "sustainable" in user_input_sustainability or "eco" in user_input_sustainability or "green" in user_input_sustainability:
    print(recommend_sustainable_crypto(crypto_db))
else:
    print("Try asking about eco-friendly or sustainable crypto options!")


```

```python
# ==============================
# Testing & Final Touches - Coded by Caroline Wanjiru
# ==============================

# Define placeholder functions for now
def handle_profit_queries(query):
    """Placeholder function to handle profit-related queries."""
    if "rising" in query or "profit" in query:
        # This is a simplified response for testing
        return "Looking for rising coins? Consider Bitcoin, Ethereum, or Solana."
    return None

def handle_sustainability_queries(query):
    """Placeholder function to handle sustainability-related queries."""
    if "green" in query or "sustainable" in query or "eco" in query:
        # This is a simplified response for testing
        return "Interested in green options? Check out Cardano and Solana."
    return None


def run_tests():
    """Automated test cases with visual output"""
    print("\n" + "="*50)
    print("üß™ RUNNING SYSTEM TESTS")
    print("="*50)

    test_cases = [
        ("Which coin is rising?", "profit", ["Bitcoin", "Ethereum", "Solana"]),
        ("Show me green coins", "sustainability", ["Cardano", "Solana"]),
        ("What should I buy?", "unknown", []),
        ("exit", "exit", [])
    ]

    for query, query_type, expected in test_cases:
        print(f"\nTEST CASE: '{query}'")
        print("-"*40)

        if query_type == "exit":
            print("‚úÖ Exit command recognized")
            continue

        # Ensure the functions are called correctly
        response = handle_profit_queries(query) if query_type == "profit" else handle_sustainability_queries(query)

        if not response:
            print("‚ö†Ô∏è No response generated" if query_type != "unknown" else "‚úÖ Correctly ignored unknown query")
        else:
            # Simplified check for now
            passed = all(coin in response for coin in expected)
            print(f"{'‚úÖ PASS' if passed else '‚ùå FAIL'} - Response: {response}")

def main():
    """Enhanced main conversation flow"""
    print("\n" + "="*50)
    print(f"üöÄ WELCOME TO {bot_name.upper()}!")
    print("="*50)
    print("Type 'examples' to see query samples or 'exit' to quit\n")

    run_tests()  # Show test results on startup

    while True:
        try:
            user_input = input("\nYour query: ").strip()

            if user_input.lower() == 'exit':
                print(f"\n{bot_name}: üîí Session ended. Always verify with CoinGecko!")
                break
            elif user_input.lower() == 'examples':
                print("\nüí° Try these queries:")
                print("- 'Which crypto is rising?'")
                print("- 'Show me sustainable coins'")
                print("- 'Compare energy usage'")
                continue

            # Process queries
            response = handle_profit_queries(user_input) or \
                      handle_sustainability_queries(user_input) or \
                      "I can analyze:\n ‚Ä¢ Profit trends üíπ\n ‚Ä¢ Sustainability ‚ôªÔ∏è\nTry: 'rising coins' or 'green cryptos'"

            print(f"\n{bot_name}: {response}")

        except Exception as e:
            print(f"\n{bot_name}: ‚ùó Error processing request - please try again")

# ==============================
# Launch the Chatbot
# ==============================
if __name__ == "__main__":
    main()
```
